```python
from hmmlearn.hmm import GaussianHMM
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./tab_features_flat/1000180 - A-Ha - The Sun Always Shines On Tv (guitar pro).gp5.csv',index_col=0)
for col in data.columns:
    if col.startswith('counts_') or col.startswith('delta') or col.startswith('source_'):
        data.drop(col,axis=1,inplace=True)
data.describe()
```

```python
n = 5
model = GaussianHMM(n_components=n)
model.means_ = [1.0/pow(2,i) for i in range(n)]
model.fit(data[['duration']])
```

```python
model.score(data[['duration']])
```
